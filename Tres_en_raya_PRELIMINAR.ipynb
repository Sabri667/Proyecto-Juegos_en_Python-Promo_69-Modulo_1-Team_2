IMPRIMIR TABLERO

```python
#IMPRIMIR TABLERO
def imprimir_tablero(tablero):
    for indice,valor in enumerate(tablero):
        print("|".join(valor))
        if indice != len(tablero)-1:
            print("-----------")
```

```python
#TABLEROS MODELO

#TABLERO VACÍO
L0=["   ","   ","   "]
tablero_inicial=[L0,L0,L0]
imprimir_tablero(tablero_inicial)
print(" ")

#TABLERO CON POSICIONES (AYUDA AL USUARIO)
L1 = [' 1 ',' 2 ',' 3 ']

L2= [' 4 ',' 5 ',' 6 ']

L3= [' 7 ',' 8 ',' 9 ']

tablero_con_posiciones=[L1,L2,L3]
imprimir_tablero(tablero_con_posiciones)

#TABLERO JUGADA
print("")
tablero_jugada=tablero_inicial.copy()
imprimir_tablero(tablero_jugada)
```

1.SALUDO Y VALIDACIÓN DE USUARIOS

```python
#INICIO DE JUEGO: VALIDACIÓN DE USUARIOS
#print("¡Hola! Bienvenidos al clásico 'Tres en raya' o 'Tic Tac Toe'.\n\nEstas son las Reglas del Juego:\n\n*El juego se juega en un tablero de 3x3 con 2(dos) jugadores, uno representa 'X', el otro 'O' y van alternando turnos.\n*El objetivo es obtener 3(tres) símbolos del mismo tipo en línea horizontal, vertical o diagonal.\n*El juego termina cuando un jugador gana o el tablero se llena (empate).\n*El jugador que logre colocar tres de sus símbolos en línea es el ganador.")

def validar_jugadores():
    while True:
        try:
            jugadores=input("Para comenzar, ingresa el nombre del jugador 1 y 2 separados por una coma (','): ").strip().replace(" ","").upper()
            jugador_1,jugador_2=jugadores.split(",")
            
            if not jugadores:
                print("¡Ups! Parece que no has escrito nada. Vuelve a intentarlo.")
                continue
            elif not jugador_1 or not jugador_2:
                continue
            else:
                print (f"¡Hola {jugador_1}, {jugador_2}!\n\nBienvenidos al clásico 'Tres en raya' o 'Tic Tac Toe'.\n\nEstas son las Reglas del Juego:\n\n*El juego se juega en un tablero de 3x3 con 2(dos) jugadores, uno representa 'X', el otro 'O' y van alternando turnos.\n*El objetivo es obtener 3(tres) símbolos del mismo tipo en línea horizontal, vertical o diagonal.\n*El juego termina cuando un jugador gana o el tablero se llena (empate).\n*El jugador que logre colocar tres de sus símbolos en línea es el ganador.")
                return
        except:
            print("Debes ingresar 2(dos) jugadores, si quieres intentarlo de nuevo, ejecuta el programa.")
            break
```
