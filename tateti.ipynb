```python
import time

# 1. Función para imprimir el tablero (formato lista de listas).
def imprimir_tablero(tablero):
    for i in range(3):
        fila = tablero[i]
        print(" | ".join(fila))
        if i < 2:
            print("---------")

# 2. Inicialización del tablero de juego (lista de listas vacías).
L1 = [" ", " ", " "]
L2 = [" ", " ", " "]
L3 = [" ", " ", " "]
tablero_partida = [L1, L2, L3]

# Tablero de referencia con números para mostrar posiciones al usuario.
L1_ref = ["1", "2", "3"]
L2_ref = ["4", "5", "6"]
L3_ref = ["7", "8", "9"]
tablero_movimientos = [L1_ref, L2_ref, L3_ref]

# Lista de posiciones disponibles (para validar movimientos)
posiciones_tablero = ["1","2","3","4","5","6","7","8","9"]

# 3. Función para realizar un movimiento con validación de casilla ocupada.
def mover(ficha, tablero, posiciones):
    while True:
        movimiento = input(f'Introduce en qué casilla quieres poner tu {ficha}: ')
        if movimiento not in posiciones:
            print("\nCasilla no disponible o inválida. Elige otra.\n")
            continue
        
        # Convierte número a índice de fila y columna (0-2).
        num = int(movimiento) - 1
        fila = num // 3
        col = num % 3
        
        if tablero[fila][col] != " ":
            print("\nCasilla ocupada. Elegí otra.\n")
            continue
        
        # Coloca la ficha y actualiza posiciones disponibles
        tablero[fila][col] = ficha
        posiciones.remove(movimiento)
        return True

# 4. Valida si hay ganador
def verificar_ganador(tablero, ficha):
    # Verifica filas.
    for i in range(3):
        if tablero[i][0] == ficha and tablero[i][1] == ficha and tablero[i][2] == ficha:
            return True
    # Verifica columnas.
    for i in range(3):
        if tablero[0][i] == ficha and tablero[1][i] == ficha and tablero[2][i] == ficha:
            return True
    # Verifica diagonales.
    if tablero[0][0] == ficha and tablero[1][1] == ficha and tablero[2][2] == ficha:
        return True
    if tablero[0][2] == ficha and tablero[1][1] == ficha and tablero[2][0] == ficha:
        return True
    return False

# 5. Valida si hay empate.
def verificar_empate(posiciones):
    return len(posiciones) == 0

# 6. Presentación del juego.
print("\n\nPara empezar, introduce los nombres de los jugadores:\n")
jugador1 = input('Nombre del Jugador 1 (X): ').title().strip() or 'Jugador 1'
jugador2 = input('Nombre del Jugador 2 (O): ').title().strip() or 'JUgador 2'

print(f"¡Hola {jugador1} y {jugador2}!\n\nBienvenidxs al clásico 'Tres en raya' o 'Tic Tac Toe'.\n")
print("Reglas del Juego:\n")
print("* El juego se juega en un tablero de 3x3 con 2 jugadores, uno representa 'X', el otro 'O' y van alternando turnos.")
print("* El objetivo es obtener 3 símbolos del mismo tipo en línea horizontal, vertical o diagonal.")
print("* El juego termina cuando un jugador gana o el tablero se llena (empate).")
print("* El jugador que logre colocar tres de sus símbolos en línea es el ganador.\n")

print("Este es el tablero inicial:\n")
imprimir_tablero(tablero_partida)

print("\nEn cada turno el jugador deberá indicar el número de celda donde marcará su jugada, siempre que esté vacía.")
print("Aquí están las posiciones:\n")
imprimir_tablero(tablero_movimientos)
print("\n¡Buena Suerte!\n")

print(f"Empieza a colocar {jugador1}:\n")

# 7. Bucle principal (alterna turnos hasta ganador o empate)
turno = 0  # 0 = jugador1 (X), 1 = jugador2 (O)
ganador = False
empate = False

while not ganador and not empate:
    if turno == 0:
        ficha = "X"
        jugador_actual = jugador1
    else:
        ficha = "O"
        jugador_actual = jugador2
    
    print(f"\nTurno de {jugador_actual} ({ficha})")
    print()
    mover(ficha, tablero_partida, posiciones_tablero)
    imprimir_tablero(tablero_partida)
    
    # Verifica si hay ganador después del movimiento
    if verificar_ganador(tablero_partida, ficha):
        print()
        print(f"\n¡Enhorabuena {jugador_actual}, HAS GANADO!")
        ganador = True
        break
    13
    # Verifica si hay empate
    if verificar_empate(posiciones_tablero):
        print()
        print(f"\n{jugador1} y {jugador2}... ¡Habéis empatado!")
        empate = True
        break
    
    # Cambia turno
    turno = 1 - turno
    time.sleep(0.5)

print("\nFin del juego.\nGracias por participar.")
```
